# Find Pairs

```python
from malariagen_data.pv4 import Pv4
import pandas as pd
```

```python
pv4 = Pv4("gs://pv4_staging/")
```

```python
pv4_metadata = pv4.sample_metadata()

pv4_metadata.head()
```

Filter to include samples in the analysis set with percent callable over 50%

```python
loc_filtered_samples = ((pv4_metadata["% callable"] > 50) 
                        & (pv4_metadata['All samples same individual'].str.contains(",")))
pv4_metadata_pairs = pv4_metadata.loc[loc_filtered_samples]
```

```python
pv4_metadata_pairs
```

Check at least two samples in the pair exist after filtering 

```python
pairs_both_included = []
samples = list(pv4_metadata_pairs.Sample)

for pairs in pv4_metadata_pairs['All samples same individual']: 
    count = 0
    for pair in pairs.split(','): 
        if pair in samples: 
            count +=1
    if count >= 2: 
        pairs_exist = True 
    else: 
        pairs_exist = False
    pairs_both_included.append(pairs_exist)
```

```python
pv4_metadata_pairs.loc[pairs_both_included]
```
